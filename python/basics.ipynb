# Print to console

Syntax of print() function:
```python
print(*objects, sep = ' ', end = '\n', file = sys.stdout, flush = False)
```

where:
- **objects** are comma separated objects that has to be printed to the file.
- **sep** is the separator used between objects. **sep** can be a string or a character. By default **sep** is space.
- **end** is the character or a string that has to appended after the **objects**.
- **file** is the place where our print function has to write the objects. By default it is **sys.stdout**, in other words system standard output which is console output.

````python
print("Hello world!")
````

# Type Casting

````python
# integer
n = 100
print(n)
print(type(n), end='\n\n')

# float
f = float(n)
print(f)
print(type(f), end='\n\n')

# string
s = str(f)
print(s)
print(type(s), end='\n\n')

# casting string to int
n = int(float(s)) # cast to float then to int
print(n)
print(type(n), end='\n\n')
````

# For Loop

The syntax of For Loop in Python is

```python
for item in iterable:
    statement(s)
```

You can access the **item** variable inside for block. `iterable` could be a sequence or collection.

## For Loop with Range

````python
for i in range(25, 29):
    print(i)
````

## For Loop with List

````python
mylist = ['python', 'programming', 'examples', 'programs']

for i in mylist:
    print(i)
````

## For Loop with Tuple

````python
mytuple = ('python', 'programming', 'examples', 'programs')

for i in mytuple:
	print(i)
````

## For Loop with Dictionary

````python
mydictionary = {'name' : 'python', 'category' : 'programming', 'topic' : 'examples'}

for x in mydictionary:
	print(x, ':', mydictionary[x])
````

## For Loop with Set

````python
myset = {'python', 'programming', 'examples'}

for x in myset:
	print(x)
````
